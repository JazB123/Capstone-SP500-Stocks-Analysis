# Exploratory Data Analysis of Stock Market Data
The following code will do a EDA of the stock market data we previously had wrangled. We will start by first importing the proper modules and the wrangled data.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import math

#Importing the DataFrame
df = pd.read_csv('WrangledData.csv', header=[0,1])

#Cleaning up the data a little, due to messy import
df = df.drop(0)
df['Date'] = pd.to_datetime(df[df.columns[0]])
df.set_index('Date', inplace=True)
df = df.drop(columns=df.columns[0])
```

# More Data Cleanup
The next section will be doing a little more data cleanup to get rid of dates with the most zero values, as these zero values will have significant effect on our data analysis.

```python
#We will call the DataFrame with missing values dfna
#We want to keep this DataFrame in case we want to do a more in-depth analysis on a specific stock
dfna = df
#the DataFrame 'df' will have a reasonable amount of NA's removed for better data analysis
df = df.T
nanperday = df.xs('open', level=1).isna().sum()
plt.plot(df.columns, nanperday)
plt.ylabel('NaN Values Per Day')
```

As can be seen in this plot, we hit a constant minimum around 2015. We will thus attempt to remove all data points up to this date and time.

```python
#Running a min gives us a value of 4
nanperday.min()
#Finding all days that equal 4 doesn't give us a very wide date range, so lets try going up by one
nanperday[nanperday <= 4]
#This looks better
nanperday[nanperday <= 5]
#Lets test this up to a certain value and see were we get diminishing returns on datapoints
for i in range(20):
    print('Total Datapoints of dates equal', len(nanperday[nanperday <= i]), 'Nan less than or equal to', i)
```

```python
#Looks like 10 is our lucky number, from the following code we can see that this date range is from about 2015 to 2022
print(nanperday[nanperday <= 10])
#Lets subset our dataframe with these values
df = df.T
df = df[nanperday <= 10]
```

```python
#Lets Transpose our data and check it over to make sure it looks good
df.T
```

```python
#Now we want to remove those 10 data points with the NA values, let's first determine what they are
nan2 = df.isna().sum() != 0
#Now we subset the Transpose of our dataset
df.T[nan2]
#From this we get a list of 13 stocks, which is weird we were only expecting 10
#This is most likely to a stock being delisted or missing data after our initial cutoff
#A glance through of our original data supports this
```

```python
#lets create a list of all these stocks
needdrop = ['ALTR', 'BHI', 'BSX', 'FRMC', 'HPE', 'IR', 'NMHLY', 'TIME', 'TW', 'UA', 'CPICQ', 'CTQ', 'SONC']
#Transpose the index back to having the Stocks as the index
df = df.T
#Now drop those values
for i in range(len(needdrop)):
    df.drop(index=needdrop[i], inplace=True)
```

```python
#Now lets check our data again
df[df.T.isna().sum() != 0]
#Looks like there are no missing values!
```

# Data Analysis
Finally we can get to the whole point of this notebook! Lets begin some analysis of our dataset. Lets first look at some correlations

```python
#Lets see if any of these stocks correlate with one another in price action
corrmatrix = df.xs('percentchange', level=1).T.corr()
#This code shows the Pearson Correlation coefficient for each stock
#This can be anywhere between 1 and -1, where 1 is complete correlation, while -1 is complete inverse correlation


#Comparing this to another source, https://www.macroaxis.com/invest/market/AAPL--compareProfile--A
#Our numbers are extremely similar to this calculator


#This is a lot to look at, lets only look at the strongly correlated stocks, not including duplicates, or 1 to 1 ratios
strongcorr = corrmatrix[(corrmatrix > 0.75) | (corrmatrix < -0.75)]
corrdf = pd.DataFrame()
colist = []
for column, series in strongcorr.items():
    if len(series.dropna()) == 0:
        continue
    x, y = zip(*series.dropna().items())
    for i in range(len(x)):
        if math.isnan(y[i]) == True:
            continue
        if any(r == x[i] for r in colist):
            continue
        tempdf = pd.DataFrame({'CorrCoef' : [y[i]], 'Stock1' : column, 'Stock2' : x[i]})
        corrdf = pd.concat([corrdf, tempdf])
        if any(r == column for r in colist):
            continue
        colist.append(column)
corrdf = corrdf[corrdf != 1].dropna()
corrdf = corrdf.sort_values(by='CorrCoef', ascending=False)
corrdf
```

To verify our data is correct, I used the online calculator mentioned in the cell above to test correlations of some of these stocks listed, and the values I calculated where extremely similar. Looking at this we can see strong correlations between stocks in the same sectors, this absolutely makes sense as sectors tend to trade together. An example of sectors would be, Technology, Finance, Agriculture, etc. 

```python
#Next we want to see if there is any correlation between volume and price action
df.xs('AAL', level=0).T.drop(columns=['close', 'open']).corr()['volume'][0]
#This code in one line gives us the Correlation coefficient between price action and volume for AAL.
#Lets make a dictionary for every stock
volactiondict = {}
allstocks = df.index.get_level_values(0).drop_duplicates()
for i in range(len(allstocks)):
    volactiondict[allstocks[i]] = df.xs(allstocks[i], level=0).T.drop(columns=['close', 'open']).corr()['volume'][0]
#Lets turn that into a DataFrame
volactiondf = pd.DataFrame(index = volactiondict.keys(), data=volactiondict.values(), columns=['Pearson Coef Vol vs PriceAction'])
volactiondf
```

```python
plt.scatter(y = volactiondf.values, x = list(range(len(volactiondf.values))))
plt.title('Pearson Coefficent Price Action vs Volume')
```

```python
#Hmm so there is no clear correlation, lets try again but only use the extreme cases of volume, when it's high
for x, y in df.xs('AAL', level=0).T.drop(columns=['close', 'open', 'percentchange']).items():
    column = x
    series = y
v, b = pd.qcut(y, q=4, retbins=True)
tempdf = df.xs('AAL', level=0).T.drop(columns=['close', 'open'])
tempdf[(tempdf['volume'] > b[3])]
#This code split the AAL stock into its 75th quantile range, so the highest 75%
#Lets put this into our previous for loop
volactiondict = {}
allstocks = df.index.get_level_values(0).drop_duplicates()
for i in range(len(allstocks)):
        for x, y in df.xs(allstocks[i], level=0).T.drop(columns=['close', 'open', 'percentchange']).items():
            series = y
        v, b = pd.qcut(y, q=4, retbins=True, duplicates='drop')
        if len(b) < 4:
            continue
        tempdf = df.xs(allstocks[i], level=0).T.drop(columns=['close', 'open'])
        tempdf = tempdf[(tempdf['volume'] > b[3])]
        volactiondict[allstocks[i]] = tempdf.corr()['volume'][0]
volactiondf = pd.DataFrame(index = volactiondict.keys(), data=volactiondict.values(), columns=['Pearson Coef Vol vs PriceAction'])
volactiondf
plt.scatter(y = volactiondf.values, x = list(range(len(volactiondf.values))))
plt.title('Pearson Coefficent Price Action vs Volume 75th Quantile')
```

```python
#Very clear strong negative correlation with high volume, lets look at low volume
volactiondict = {}
allstocks = df.index.get_level_values(0).drop_duplicates()
for i in range(len(allstocks)):
        for x, y in df.xs(allstocks[i], level=0).T.drop(columns=['close', 'open', 'percentchange']).items():
            series = y
        v, b = pd.qcut(y, q=4, retbins=True, duplicates='drop')
        if len(b) < 4:
            continue
        tempdf = df.xs(allstocks[i], level=0).T.drop(columns=['close', 'open'])
        tempdf = tempdf[(tempdf['volume'] < b[1])]
        volactiondict[allstocks[i]] = tempdf.corr()['volume'][0]
volactiondf = pd.DataFrame(index = volactiondict.keys(), data=volactiondict.values(), columns=['Pearson Coef Vol vs PriceAction'])
volactiondf
plt.scatter(y = volactiondf.values, x = list(range(len(volactiondf.values))))
plt.title('Pearson Coefficent Price Action vs Volume 25th Quantile')
```
